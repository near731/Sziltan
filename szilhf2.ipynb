```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import sympy as sp
import math
from sympy import init_printing
import scipy.linalg as la
from sympy import *
from sympy import Matrix, symbols, solve_linear_system
from sympy.plotting import plot

print("Szilárdságtan HF2 megoldó\n")

L,h,d,F,M,p,εa,εb,εc,alfa,E,v,Iy,C11,C12,C21,C22,C31,C32=sp.symbols("L,h,d,F,M,p,εa,εb,εc,alfa,E,v,Iy,C11,C12,C21,C22,C31,C32")

#ADATOK

L_adat=1.5 #m
h_adat=2 #m
d_adat=35 #mm
F_adat=4.5 #kN
M_adat=2 #kNm
p_adat=2.5 #kN/m
εa_adat=-1.1*10**-4
εb_adat=-4.5*10**-4
εc_adat=2*10**-4
alfa_adat=2*np.pi/3 #rad
E_adat=210 #GPa
v_adat=0.3
Iy_adat=7.78*10**-7 #m4

adatok=[(L,L_adat),(h,h_adat),(d,d_adat),(F,F_adat),(M,M_adat),(p,p_adat),(εa,εa_adat),(εb,εb_adat),(εc,εc_adat),(alfa,alfa_adat),(E,E_adat),(v,v_adat),(Iy,Iy_adat)]

#1.feladatrész

print("1.feladatrész: \n")

Az=(-M-2*L*F-(p/2)*L**2)/L
Azn=Az.subs(adatok)
print("Az=",Azn,"kN")

Cz=L*p-F-Az
Czn=Cz.subs(adatok)
print("Cz=",Czn,"kN")

Bz=Czn
print("Bz=",Bz,"kN")

#IGÉNYBEVÉTELI FÜGGVÉNYEK

print("Igénybevételi függvény:\n")

x=sp.symbols("x")

Mh=sp.Piecewise((-F*x-M, x<=L),(-F*x-M-Az*(x-L), x<=2*L),((p/2)*(3*L-x)**2, x<=3*L))
Mhn=Mh.subs(adatok)
print(Mhn)

#IGÉNYBEVÉTELI ÁBRA

import matplotlib.pyplot as plt
from numpy import linspace

print("\nIgénybevételi ábrák:\n")

L0=(3*L).subs(adatok)
xs=linspace(0.,float(L0),500)

Mhxs=[Mhn.subs(x,xi) for xi in xs]

plt.plot(xs, Mhxs, lw=3)
plt.xlabel("$x$ (m)")
plt.ylabel("$Mh(x)$ (kNm)")
plt.grid()
plt.show()

Mh0=-F*0-M
Mh0n=Mh0.subs(adatok)
print("Mh(0)=",Mh0n,"kNm")

MhL=-F*L-M
MhLn=MhL.subs(adatok)
print("Mh(L)=",MhLn,"kNm")

Mh2L=-F*2*L-M-Az*(2*L-L)
Mh2Ln=Mh2L.subs(adatok)
print("Mh(2L)=",Mh2Ln,"kNm")

Mh3L=(p/2)*(3*L-3*L)**2
Mh3Ln=Mh3L.subs(adatok)
print("Mh(3L)=",Mh3Ln,"kNm")

#2.feladatrész

print("\n2.feladatrész:\n")

Mh=(p/2)*(3*L-x)**2
mh= Mh.diff(x)
m=mh.subs(adatok)
print("Mh' = ",m)
print("Érintő meredeksége",2*L," helyen: ",m.subs({x:2*L_adat}))
print("Érintő meredeksége ",3*L," helyen: ",m.subs({x:3*L_adat}),"\n")

print("Iy=",Iy_adat,"mm4")
print("E=",E_adat,"GPa")

print("\n Peremfeltételek:\n")
print("PF1: w1(L)=0")
print("PF2: w2(L)=0")
print("PF3: w2(2L)=0")
print("PF3: w2(2L)=0")
print("PF4: w3(2L)=0")
print("PF5: w1'(L)=w2'(L)")
print("PF5: w2'(2L)=w3'(2L)")

MH_1 = -F*x-M
MH_2 = -F*x-M-Az*(x-L)
MH_3 = (p/2)*(3*L-x)**2


w_1 = MH_1 #konstans 1000 szorzó az SI miatt
w_2 = MH_2 
w_3 = MH_3 

MH_1_1 =  (w_1.integrate(x) + C11)
MH_1_2 = (MH_1_1.integrate(x) + C12)
MH_2_1 = (w_2.integrate(x) +C21)
MH_2_2 = (MH_2_1.integrate(x) +C22)
MH_3_1 = (w_3.integrate(x) + C31)
MH_3_2 = (MH_3_1.integrate(x) +C32)

MH_1_1n=MH_1_1.subs(adatok)
MH_1_2n=MH_1_2.subs(adatok)
MH_2_1n=MH_2_1.subs(adatok)
MH_2_2n=MH_2_2.subs(adatok)
MH_3_1n=MH_3_1.subs(adatok)
MH_3_2n=MH_3_2.subs(adatok)

print("\n\nMh_1 első integrálja: ",MH_1_1n)
print("Mh_1 második integrálja: ",MH_1_2n)

print("\nMh_2 első integrálja: ",MH_2_1n)
print("Mh_2 második integrálja: ",MH_2_2n)

print("\nMh_3 első integrálja: ",MH_3_1n)
print("Mh_3 második integrálja: ",MH_3_2n)

print("\nKonstansok meghatározása:")

w_1_2L=MH_2_2n.subs({x:2*L_adat})
w_1_L=MH_2_2n.subs({x:L_adat})

print(w_1_2L-w_1_L,"=0")
C21n=(w_1_2L-w_1_L-1.5*C21)/-1.5
print("\nC21=",C21n,"\n")

print(w_1_L.subs({C21:C21n}),"=0")
C22n=(w_1_L.subs({C21:C21n})-C22)/-1
print("\nC22=",C22n,"\n")

C11n=1.62
print("C11=",C11n)

C12n=2.35125
print("C12=",C12n)

C31n=-37.33265
print("C31=",C31n)

C32n=31.2792
print("C32=",C32n)

print("\nDifferenciálegyenletek végleges alakja:")
K=(Iy*E*10**6)
Kn=K.subs(adatok)

print("Hajlitómerevség:",Kn)
w1_x=(-1/K)*(MH_1_2n.subs({C11:C11n},{C12:C12n}))
w1_xn=w1_x.subs(adatok)
print("w1_x=",w1_xn)

#LEHAJLÁS FÜGGVÉNY
print("\n\nLehajlás függvény:\n")
w1=(-1/K)*(-0.75*x**3-x**2+11.7457125*x-12.83731875)
w1n=w1.subs(adatok)
print("w1_x=",w1n)

w2=(-1/K)*(1.284*x**3-10.15625*x**2+25.480125*x-19.7021)
w2n=w2.subs(adatok)
print("w2_x=",w2n)

w3=(-1/K)*(5/48*x**4-15/8*x**3+12.65625*x**2-37.332525*x+40.2792)
w3n=w3.subs(adatok)
print("w3_x=",w3n)

w=sp.Piecewise((w1*1000, x<=L),(w2*1000, x<=2*L),(w3*1000, x<=3*L))
wn=w.subs(adatok)
print(wn)

L0=(3*L).subs(adatok)
xs=linspace(0.,float(L0),500)

wxs=[wn.subs(x,xi) for xi in xs]

plt.plot(xs, wxs, lw=3)
plt.xlabel("$x$ (m)")
plt.ylabel("$w(x)$ (mm)")
plt.grid()
plt.show()

w1_0=(-1/163.38)*(-0.75*0**3-0**2+11.7457125*0-12.83731875)
print("\nwmax=",w1_0*1000,"mm")

w1_L=(-1/163.38)*(-0.75*(L)**3-L**2+11.7457125*(L)-12.83731875)
w1_Ln=w1_L.subs(adatok)
print("w1_L=",w1_Ln*1000,"mm")

w2_L=(-1/163.38)*(1.284*(L)**3-10.15625*(L)**2+25.480125*(L)-19.7021)
w2_Ln=w2_L.subs(adatok)
print("w2_L=",w2_Ln*1000,"mm")

w2_2L=(-1/163.38)*(1.284*(2*L)**3-10.15625*(2*L)**2+25.480125*(2*L)-19.7021)
w2_2Ln=w2_2L.subs(adatok)
print("w2_2L=",w2_2Ln*1000,"mm")

w3_2L=(-1/163.38)*(5/48*(2*L)**4-15/8*(2*L)**3+12.65625*(2*L)**2-37.332525*(2*L)+40.2792)
w3_2Ln=w3_2L.subs(adatok)
print("w3_2L=",w3_2Ln*1000,"mm")

w3_3L=(-1/163.38)*(5/48*(3*L)**4-15/8*(3*L)**3+12.65625*(3*L)**2-37.332525*(3*L)+40.2792)
w3_3Ln=w3_3L.subs(adatok)
print("w3_3L=",w3_3Ln*1000,"mm")

#SZÖGELFORDULÁS

print("\nSzögelfordulás:\n")

φI=(-1/K)*(-2.25*x**2-2*x+11.7457125)
φIn=φI.subs(adatok)
print("φI=",φIn)

φII=(-1/K)*(3.85415*x**2-20.3125*x+25.480125)
φIIn=φII.subs(adatok)
print("φII=",φIIn)

φIII=(-1/K)*(5/12*x**3-45/8*x**2+25.3125*x-37.332525)
φIIIn=φIII.subs(adatok)
print("φIII=",φIIIn)

φ=sp.Piecewise((φI, x<=L),(φII, x<=2*L),(φIII, x<=3*L))
φn=φ.subs(adatok)
print("φ=",φn)

φxs=[φn.subs(x,xi) for xi in xs]

plt.plot(xs, φxs, lw=3)
plt.xlabel("$x$ (m)")
plt.ylabel("$φ(x)$ (rad)")
plt.grid()
plt.show()

φI_0=(-1/163.38)*(-2.25*0**2-2*0+11.7457125)
print("φmax=",φI_0,"rad")

φI_L=(-1/163.38)*(-2.25*L**2-2*L+11.7457125)
φI_Ln=φI_L.subs(adatok)
print("φ1_L=",φI_Ln,"rad")

φII_L=(-1/163.38)*(3.85415*L**2-20.3125*L+25.480125)
φII_Ln=φII_L.subs(adatok)
print("φ2_L=",φII_Ln,"rad")

φII_2L=(-1/163.38)*(3.85415*(2*L)**2-20.3125*(2*L)+25.480125)
φII_2Ln=φII_2L.subs(adatok)
print("φ2_2L=",φII_2Ln,"rad")

φIII_2L=(-1/163.38)*(5/12*(2*L)**3-45/8*(2*L)**2+25.3125*(2*L)-37.332525)
φIII_2Ln=φII_2L.subs(adatok)
print("φ3_2L=",φIII_2Ln,"rad")

φIII_3L=(-1/163.38)*(5/12*(3*L)**3-45/8*(3*L)**2+25.3125*(3*L)-37.332525)
φIII_3Ln=φIII_3L.subs(adatok)
print("φ3_3L=",φIII_3Ln,"rad")

#3.feladatrész

print("\n3.feladatrész: \n")

σF=240 #Mpa
print("Folyáshatár:",σF,"MPa")

λ0=105
print("λ0=",λ0)

λ=sp.symbols("λ")

σKr=308-1.14*λ
print("σKr=",σKr,"MPa")

λF=(-σF+308)/1.14
print("\nλF=",λF)

σ0=308-1.14*λ.subs({λ:λ0})
print("σ0=",σ0,"MPa")

σE=E*10**3*(np.pi/λ)**2

σλ=sp.Piecewise((σF, λ<=λF),(308-1.14*λ, λ<=λ0),(σE, λ>λ0))
σλn=σλ.subs(adatok)
print("σλ=",σλn)

L0=210
λs=linspace(0.,float(L0),500)

σλs=[σλn.subs(λ,λi) for λi in λs]

plt.plot(λs, σλs, 'r',lw=3)
plt.xlabel("$λ$")
plt.ylabel("$σ$ (MPa)")
plt.grid()
plt.show()

print("\nMéretezés:")

Ft=3*Cz*1000
Ftn=Ft.subs(adatok)
print("\nFt=",Ftn,"N")

t=sp.symbols("t")

T=(d-(d**4-(64*Ft)/(np.pi*E*1000*(np.pi/(h*1000))**2))**0.25)/2
Tn=T.subs(adatok)
print("\nt=",Tn,"mm")

t=round(Tn,1)
print("\nt=",t,"mm")

I_2=(np.pi*(d**4-(d-2*t)**4))/64
I_2n=I_2.subs(adatok)
print("\nI_2=",I_2n,"mm4")

A=((d**2-(d-2*t)**2)*np.pi)/4
An=A.subs(adatok)
print("A=",An,"mm2")

i2=(I_2/A)**0.5
i2n=i2.subs(adatok)
print("i2=",i2n)

λ=(h*1000)/i2
λn=λ.subs(adatok)
print("λ=",λn)

#4.feladatrész

print("\n4.feladatrész:\n")

εx=εa
εxn=εx.subs(adatok)
print("εx=",εxn*10**4,"*10^-4")

εz=εc
εzn=εz.subs(adatok)
print("εz=",εzn*10**4,"*10^-4")

γxz=2*(εb-εx*(np.cos(2*np.pi/3))**2-εz*(np.sin(2*np.pi/3))**2)/np.sin(2*2*np.pi/3)
γxzn=γxz.subs(adatok)
print("\nγxz=",γxzn*10**4,"*10^-4")

εy=(-v/(1-v))*(εx+εz)
εyn=εy.subs(adatok)
print("εy=",εyn*10**4,"*10^-4")

εv=εx+εy+εz
εvn=εv.subs(adatok)
print("εv=",εvn*10**4,"*10^-4")

ε=[[εx,0,γxz],
   [0, εy, 0],
   [γxz,0,εz]]

εn=[[εxn*10**4,0,1/2*γxzn*10**4],
   [0, εyn*10**4, 0],
   [1/2*γxzn*10**4,0,εzn*10**4]]

print("\nεn=",εn)
    
σx=E*1000/(1+v)*(εx+v/(1-2*v)*εv)
σxn=σx.subs(adatok)
print("\nσx=",σxn,"MPa")

σy=0
print("σy=0 MPa")

σz=E*1000/(1+v)*(εz+v/(1-2*v)*εv)
σzn=σz.subs(adatok)
print("σz=",σzn,"MPa")

Txz=E*1000/(1+v)*1/2*γxz
Txzn=Txz.subs(adatok)
print("Txz=",Txzn,"MPa")

σ=[[σxn,0,Txzn],
   [0, σy, 0],
   [Txzn,0,σzn]]

print("\nσ=",σ)

print("\nInvariáns mennyiségek:\n")

σI=σx+σy+σz
σIn=σI.subs(adatok)
print("σI=",σIn,"MPa")

σII=σx*σy+σx*σz+σy*σz-Txz**2
σIIn=σII.subs(adatok)
print("σII=",σIIn,"MPa")

σIII=0
print("σIII=",σIII,"MPa")

#5.feladatrész

print("\n5.feladatrész:\n")
print("Főfeszültségek meghatározása:\n")

σK=(σx+σz)/2
σKn=σK.subs(adatok)
print("σK=",σKn,"MPa")

R=(((σx-σz)/2)**2+Txz**2)**(1/2)
Rn=R.subs(adatok)
print("R=",Rn,"MPa")

σ_I=σK+R
σ_In=σ_I.subs(adatok)
print("σ_I=",σ_In,"MPa")

σ_II=0
print("σ_II=",σ_II,"MPa")

σ_III=σK-R
σ_IIIn=σ_III.subs(adatok)
print("σ_III=",σ_IIIn,"MPa")

print("\nFőirányok meghatározása:\n")

fi_I=(σ_I-σx)/Txz
fi_In=fi_I.subs(adatok)

φ_I=np.arctan(1.26159003937109)*180/np.pi
print("φ_I=",φ_I,"fok")

φ_III=φ_I+90
print("φ_III=",φ_III,"fok")



n_I=[[np.cos(φ_I*np.pi/180)],
   [0],
   [np.sin(φ_I*np.pi/180)]]

print("\nn_I=",n_I)

n_II=[[0],
      [1],
      [0]]

print("n_II=",n_II)

n_III=[[np.cos(φ_III*np.pi/180)],
   [0],
   [np.sin(φ_III*np.pi/180)]]

print("n_III=",n_III)

print("\nEllenőrzés:")

#6.feladatrész

print("\n6.feladatrész:\n")

σe_Mohr=σ_I-σ_III
σe_Mohrn=σe_Mohr.subs(adatok)
print("σe_Mohr=",σe_Mohrn,"MPa")

σe_HMH=(0.5*((σ_I-σ_II)**2+(σ_I-σ_III)**2+(σ_II-σ_III)**2))**0.5
σe_HMHn=σe_HMH.subs(adatok)
print("σe_HMH=",σe_HMHn,"MPa")

Δσe=σe_Mohr-σe_HMH
Δσen=Δσe.subs(adatok)
print("Δσe=",Δσen,"MPa")

#7.feladatrész

print("\n7.feladatrész:\n")

u=0.5*(σx*εx+2*Txz*0.5*γxz+σz*εz)
un=u.subs(adatok)
print("u=",un,"J/cm3")

u_h=1/6*σI*εv
u_hn=u_h.subs(adatok)
print("u_h=",u_hn,"J/cm3")

u_d=u-u_h
u_dn=u_d.subs(adatok)
print("u_d=",u_dn,"J/cm3")
```
